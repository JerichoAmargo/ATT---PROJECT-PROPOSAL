# Prevalence of certain health indicators among different population groups in the Philippines for various years.

````python

# Data
population_groups = ["6 months to 5 years old", "Pregnant", "Lactating", "60 years old and up"]
sub_indicators = ["Vitamin A Deficient", "Anemia"]
years = ["2003", "2008", "2011", "2013", "2014", "2015", "2019", "2020"]
data = [
    [40.1, 17.5, 20.1, np.nan],  # Vitamin A Deficient
    [32.4, 43.9, 42.2, np.nan],  # Anemia
]

# Plotting
width = 0.35
x = np.arange(len(population_groups))

fig, ax = plt.subplots(figsize=(20, 6))
bars = []
colors = ["#FF8C00", "#FFD700"]  # Color for each sub-indicator

for i in range(len(sub_indicators)):
    bars.append(ax.bar(x + (i * width), data[i], width, label=sub_indicators[i], color=colors[i]))

ax.set_title('Prevalence of Health Indicators by Population Group (Philippines)')
ax.set_xlabel('Population Group')
ax.set_ylabel('Prevalence (%)')
ax.set_xticks(x + (len(sub_indicators) * width) / 2)  # Adjust x-axis ticks
ax.set_xticklabels(population_groups)
ax.legend()

# Add data labels
for bar in bars:
    for rect in bar:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom')
        
# Move legend to the left side
plt.legend(fontsize='small', loc='center left', bbox_to_anchor=(1, 0.5))

plt.grid(True)
plt.show()
````

# Area harvested and the quantities of various fertilizers applied in different regions of the Philippines from 2003 to 2020

````python
# Create a sample DataFrame
data = {'Year': [2017, 2018, 2019, 2020, 2021],
        'Area Harvested': [100, 120, 110, 130, 150],
        'Area Applied': [50, 60, 70, 80, 90],
        'Avg Quantity Applied': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

# Calculate the growth rates
df['Area Harvested Growth Rate'] = df['Area Harvested'].pct_change() * 100
df['Area Applied Growth Rate'] = df['Area Applied'].pct_change() * 100
df['Avg Quantity Applied Growth Rate'] = df['Avg Quantity Applied'].pct_change() * 100

# Plot the growth rates
plt.figure(figsize=(20, 9))  # Set the figure size
plt.plot(df['Year'][1:], df['Area Harvested Growth Rate'][1:], label='Area Harvested')
plt.plot(df['Year'][1:], df['Area Applied Growth Rate'][1:], label='Area Applied')
plt.plot(df['Year'][1:], df['Avg Quantity Applied Growth Rate'][1:], label='Avg Quantity Applied')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.title('Growth Rates Over Time')
plt.legend()
plt.show()
````
